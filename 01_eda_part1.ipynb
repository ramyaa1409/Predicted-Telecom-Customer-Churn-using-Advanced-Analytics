```julia
using DataFrames

# Assuming df is your DataFrame containing telecom data
dfcl = DataFrame(
    customer_id = 1:5,
    payment_method = ["Bank transfer (automatic)", "Credit card (automatic)", "Mailed check", "Electronic check", "Bank transfer (automatic)"]
)

# Define a function to replace the values
function replace_payment_method(method::String)
    if method == "Mailed check" || method == "Electronic check"
        return method
    elseif method == "Bank transfer (automatic)"
        return "Bank transfer"
    elseif method == "Credit card (automatic)"
        return "Credit card"
    else
        return method
    end
end

# Apply the function to the payment_method column
dfcl.payment_method = replace_payment_method.(dfcl.payment_method)

# Display the modified DataFrame
println(dfcl)
```

```julia
using Plots
# Plot a bar chart for payment_method 
variable_name = :payment_method  
data = groupby(dfcl, variable_name)
counts = combine(data, variable_name => length => :count)
Plots.bar(counts[!, variable_name], counts[!, :count], legend=false, color="#C67812", xlabel=String(variable_name), ylabel="Count", title="Distribution of Payment Method")
```

```julia
using CSV, DataFrames, Plots, Statistics, Colors
gr()

# Select the numerical columns 
numerical_cols = select(dfbl, [:tenure_months, :monthly_charges, :total_charges, :cltv, :churn_value])

# Calculate the correlation matrix
corr_matrix = cor(Matrix(numerical_cols))

# Create a custom colormap using the specified color #C67812
custom_color = parse(Colorant, "#C67812")
custom_colormap = [colorant"white", custom_color] # Gradient from white to the specified color

# Plot the heatmap with colormap and slant text
h=heatmap(corr_matrix, color=custom_colormap, aspect_ratio=:auto,
        xticks=(1:length(names(numerical_cols)), names(numerical_cols)),
        yticks=(1:length(names(numerical_cols)), names(numerical_cols)),
        title="Heatmap of Correlation to Churn Value",
        xrotation=45, yrotation=45)

# Save the plot
savefig("correlation_heatmap_custom_color.png")

display(h)
```

```julia
using CSV, DataFrames, Plots, StatsBase  # Added StatsBase here

# Load the dataset
df = CSV.read("telco.csv", DataFrame)

# Define the mapping as before
churn_reason_map = Dict(
    "Competitor had better devices" => "competitor",
    "Competitor made better offer" => "competitor",
    "Competitor offered higher download speeds" => "competitor",
    "Competitor offered more data" => "competitor",
    "Moved" => "competitor",
    "Attitude of service provider" => "poor service",
    "Attitude of support person" => "poor service",
    "Poor expertise of online support" => "poor service",
    "Poor expertise of phone support" => "poor service",
    "Lack of self-service on Website" => "poor service",
    "Limited range of services" => "poor service",
    "Service dissatisfaction" => "poor service",
    "Extra data charges" => "expensive",
    "Lack of affordable download/upload speed" => "expensive",
    "Long distance charges" => "expensive",
    "Price too high" => "expensive"
)

# Use the `get` function with broadcasting to replace each element in the column
# This also handles cases where the churn reason might not be in the map (leaving them unchanged)
df[!, :churn_category] = get.(Ref(churn_reason_map), df[!, :churn_reason], df[!, :churn_reason])

# Remove rows where churn_category is still missing
df_filtered = dropmissing(df, :churn_category)

# Count the occurrences of each new category
category_counts = countmap(df_filtered[!, :churn_category])

# Prepare data for plotting
categories = collect(keys(category_counts))
counts = collect(values(category_counts))

# Plotting
bar(categories, counts, color="#c67812", legend=false,
    ylabel="Count", xlabel="Churn Reason", title="Churn Reasons Categorized", xrotation=45)

# Display the plot
```
